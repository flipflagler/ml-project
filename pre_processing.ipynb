```python
import numpy as np
import pandas as pd
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df=pd.read_csv("/content/car_dataset.data",names=headers)
df
```

```python
df.dtypes
```

```python
df.replace("?",np.nan,inplace=True)
df
```

```python
df.dtypes
```

```python
df.describe(include="all")
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
avg_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
print("avg of normalized-losses:",avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
```

```python
df
```

```python
df['num-of-doors'].replace(np.nan,'four',inplace=True)
df
```

```python
avg_norm_loss = df["bore"].astype('float').mean(axis=0)
print("avg of bore:",avg_norm_loss)
```

```python
df["bore"].replace(np.nan,avg_norm_loss,inplace=True)
df
```

```python
avg_norm_loss = df["stroke"].astype('float').mean(axis=0)
print("avg of stroke:",avg_norm_loss)
```

```python
df["stroke"].replace(np.nan,avg_norm_loss,inplace=True)
df
```

```python
avg_norm_loss = df["horsepower"].astype('float').mean(axis=0)
print("avg of horsepower:",avg_norm_loss)
```

```python
df["horsepower"] = df["horsepower"].replace("?", np.nan)
df["horsepower"] = df["horsepower"].astype(float)

%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(df["horsepower"])
plt.xlabel("horsepower")
plt.ylabel("count")
plt.title("horsepower bins")
plt.show()
```

```python
bins = np.linspace(min(df["horsepower"]),max(df["horsepower"]),4)
bins
```

```python
group_names = ['low','medium','high']
```

```python
df['horsepower-binned'] = pd.cut(df['horsepower'],bins,labels=group_names,include_lowest=True)
df[['horsepower' , 'horsepower-binned']].head(20)
```
